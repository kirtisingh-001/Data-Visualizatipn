```python
# Import the SparkSession class which is the "control center" for your Spark application.
from pyspark.sql import SparkSession

# Initialize the SparkSession.
# .builder starts the configuration process.
# .appName("PySpark Demo") gives your job a name for monitoring purposes.
# .getOrCreate() starts the session or connects to an existing one if it's already open.
spark = SparkSession.builder \
    .appName("PySpark Demo") \
    .getOrCreate()
```

```python
# TECHNIQUES (IMPLEMENTATION APPROACH)
# Create RDD
# Create a standard Python list (resides in your computer's local RAM).
data = [1, 2, 3, 4, 5]
# Convert the local list into an RDD (Resilient Distributed Dataset).
# This "parallelizes" the data, meaning Spark can now split it across multiple CPU cores or machines.
rdd = spark.sparkContext.parallelize(data)

print(rdd.collect())
```

```python
# Transformation: .map() applies a function (multiply by 2) to every item in the RDD.
# rdd2 doesn't actually contain the numbers yet; it contains the "plan" to double them.
rdd2 = rdd.map(lambda x: x * 2)

```

```python
# Technique 3: Actions (Execution Trigger)

# Actions trigger execution:
# Action: .collect() returns the entire dataset as a list to the driver.
rdd.collect()
print(rdd2.collect())
```

```python
# Technique 4: Common Transformations
# map()
rdd.map(lambda x: x + 10)

```

```python
# Transformation: .filter() creates a new RDD keeping only values greater than 2.
rdd.filter(lambda x: x > 2)
```

```python

# flatMap()

rdd = spark.sparkContext.parallelize(["hello world", "spark is fast"])
rdd.flatMap(lambda x: x.split(" ")).collect()
```

```python
# Technique 5: Common Actions
# collect()
rdd.collect()
```

```python
# count()
# Action: .count() calculates the total number of elements in the RDD.
rdd.count()
```

```python
# Action: .reduce() combines all elements using a specific logic (here, adding/concatenating them).
rdd.reduce(lambda x, y: x + y)
```

```python
# Imagine you are the manager of a massive global logistics company that delivers packages worldwide.

# The Problem Statement
# Every day, your company receives millions of delivery weight logs from thousands of different cities. You need to double the fuel allocation for every shipment because of a new global transport tax, and then you need a final list of these updated values to calculate your budget.

# The Challenge:

# The Data is too big: A single computer cannot process millions of logs quickly; it would take hours or crash.

# Speed: You need the results now, not tomorrow.

# 1. Start session
spark = SparkSession.builder.appName("Demo").getOrCreate()
# Instead of trying to do the work yourself, you hire a Manager
#  (SparkSession) to coordinate a whole team of workers (a cluster of computers).
# 2. Define Data
data = [10, 20, 30, 40]
# These represent the weights of your shipments (imagine these are millions of entries instead of just four).

# 3. Create RDD (Distribute Data)
rdd = spark.sparkContext.parallelize(data)
# Spark’s Solution: The manager splits the list of millions of logs and hands small chunks to different workers. Now,
# instead of one person reading a million lines, 100 workers are each reading 10,000 lines simultaneously (Parallel Processing).

# 4. Transformation: Prepare the "Double the numbers" plan (Lazy)
rdd_transformed = rdd.map(lambda x: x * 2)
# Spark’s Solution: You tell all workers: "Double the number on every log you are holding."
# Because Spark is Lazy, the workers don't start yet; they just write the instruction down.
#  This prevents them from wasting energy until you are sure the plan is final.
# 5. Action: Execute the plan and bring the result back
result = rdd_transformed.collect()
# Spark’s Solution: You give the "Go" signal.
#  All workers double their numbers at the exact same time and send their finished results back to the manager.

# 6. Output the result: [20, 40, 60, 80]
print(result)
# You get your final budget numbers instantly because the work was shared across a massive team.
```

```python
# # Scenario: Global Streaming Platform – Doubling Bandwidth Allocation
# # The Problem Statement
# # Every day, your platform receives millions of video stream logs from users worldwide.
#  Due to a new regulation, you must double the bandwidth allocation for every stream to ensure smooth playback.
#   Finally, you need a consolidated list of updated bandwidth values to calculate infrastructure costs.

# # The Challenge
# # Data Size: Millions of logs are too large for a single machine.

# # Speed: You need results instantly to adjust server capacity.
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("BandwidthProcessing").getOrCreate()
#Sample data
data = [
    (1, "UserA", 500),
    (2, "UserB", 800),
    (3, "UserC", 1200),
    (4, "UserD", 300)
]
columns = ["stream_id", "user", "bandwidth"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col
updated_df = df.withColumn("bandwidth", col("bandwidth") * 2)
updated_df.show()
```
